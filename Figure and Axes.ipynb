# Creating a Figure object

- Create our very first matplotlib plot, and
- Introduce the matplotlib Figure class.

```python
## Import the pyplot API
import matplotlib.pyplot as plt
```

```python
# First matplotlib figure

plt.figure()
plt.plot([-10,10], [-10,10])
plt.show()
```

## The Figure class

When making a matplotlib graphic the Figure class can be thought of as the container that holds all of the plot elements. It is a somewhat abstract concept and it is difficult to visualize in a jupyter notebook.

### Creating a Figure object

You can create a `Figure` object by calling plt.figure().

```python
## This creates an empty Figure object

plt.figure()
plt.show()
```

Before moving on we will demonstrate some of the most commonly used arguments to the plt.figure command, https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.figure.html.

- num: this specifies a unique identifier for the Figure object, it has to be a positive integer or a string.
- figsize: this specifies the width and height of the Figure in inches (note that jupyter notebook displays do not strictly follow the exact width and height given, but this is the size of the Figure when you save the Figure object).
- dpi: this specifies the resolution of what is plotted on the Figure and is the "dots per inch". The default is 100. Smaller values give a lower resolution image, larger values give a higher resolution.

```python
## Changing the Figure size to 8 by 6

plt.figure(figsize=(8,6))
plt.plot([-10,10], [-10,10])
plt.show()
```

```python
## A smaller dpi, dpi=50

plt.figure(figsize=(8,6),
          dpi=50)
plt.plot([-10,10], [-10,10])
plt.show()
```

```python
## A larger dpi, dpi=200

plt.figure(figsize=(8,6),
          dpi=200)
plt.plot([-10,10], [-10,10])
plt.show()
```

# The Axes Objects

- Return to our first `matplotlib` plot
- Introduce the `Axes` class
- Demonstrate `ax.plot`

## The `Axes` class

```python
# Initiate a plot
plt.figure()

# Create a line starting from (-10,10) to (10,10)
# This creates an axes object and the plot the line
plt.plot([-10,10], [-10,10])

# Add additional plot to the existing axes
plt.plot([10,-10], [-10,10])

# Show figure
plt.show()
```

## Manually adding an empty `Axes` object

While the automatic addition of an Axes object to an empty Figure object is wonderful functionality from matplotlib, we sometimes want to be able to manually add an Axes object. One way to do this is with add_subplot, https://matplotlib.org/stable/api/figure_api.html#matplotlib.figure.Figure.add_subplot.

```python
# This creates the empty figure and 
# stores it in a variable called fig
fig = plt.figure()

# this adds an Axes object to the Figure 
# and stores it in a variable called ax
ax = fig.add_subplot()

# displays the most recently created Figure object
plt.show()
```

```python
# creating Figure object and Axes object in variables
# allows us to access them later in our notebook

# ax.plot works the same way as plt.plot
ax.plot([-10,10], [-10,10])

# call fig instead of plt.show()
fig
```

## Other ways to add `Axes`

We should note that there are other ways to add `Axes` objects:

- `fig.add_axes(rect)`, https://matplotlib.org/stable/api/figure_api.html#matplotlib.figure.Figure.add_axes, and
- `fig.add_subfigure`, https://matplotlib.org/stable/api/figure_api.html#matplotlib.figure.Figure.add_subfigure.

Both of which give you slightly more control over the Axes subfigure than add_subplot. For example, we can use add_axes to add multiple Axes objects at differing locations and of differing sizes in the Figure.

```python
# Creating a fresh, empty Figure object
fig = plt.figure(figsize=(6,3))

# Adding a first Axes object
ax1 = fig.add_axes([0,0,0.25,0.25])
# Annotating it
ax1.text(.4,.4,'Axes 1')

# Adding a second Axes object
ax2 = fig.add_axes([.3,.8,.1,.1])
ax2.text(.15,.4,'Axes 2')

# Adding a third Axes object
ax3 = fig.add_axes([.7,.2,.15,.15])
ax3.text(.3,.4,'Axes3')

# Display figure
plt.show()
```

## `Axes` attributes

So what all do you get with an `Axes` object? Quite a bit! To see it all check out the documentation, https://matplotlib.org/stable/api/axes_api.html. For this notebook I will review some of the most commonly adjusted attributes:

- The actual axes we see on the plot,
- The outline of the rectangle,
- The background of the rectangle,
- The axis tick marks,
- The axis tick mark labels,
- The plot's title,
- The axis titles, and
- More.

Over the remainder of these notes we will learn various techniques for adjusting some of these elements, for example set axis titles:

```python
plt.figure()

plt.plot([-10,10],[-10,10])

# One way is with xlabel and ylabel
plt.xlabel('$x$')
plt.ylabel('$y$')

plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot()

ax.plot([-10,10],[-10,10])

# Another way is to use the Axes object
ax.set_xlabel('$x$')
ax.set_ylabel('$y$')

plt.show()
```
